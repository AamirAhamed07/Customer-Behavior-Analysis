```python
import pandas as pd

df = pd.read_csv('customers_shopping_behavior.csv') 

df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
df.head()
```

```python
#create a columns age_group
labels = ['young adult','adult','middle aged','senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[['age','age_group']].head(20)
```

```python
#create columns purchase_freaquency_day

frequency_mapping = { 'Fortnightly': 14,
                      'Weekly': 7,
                      'Monthly': 30,
                      'Quarterly': 90,
                      'Bi-Weekly': 14,
                      'Annually': 365,
                      'Every 3 Months': 90 }

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

df[['purchase_frequency_days','frequency_of_purchases']].head(15)

```

```python
df[['discount_applied','promo_code_used']].head(10)
```
